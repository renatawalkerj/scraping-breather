# Combining Correlations

## Prices correlations
To have a better perspective of the correlations, lets do it for more locations and compare:

```python
sf_dataset = dataset[dataset['location']=='san-francisco']
bt_dataset = dataset[dataset['location']=='boston']
ch_dataset = dataset[dataset['location']=='chicago']
ln_dataset = dataset[dataset['location']=='london']
la_dataset = dataset[dataset['location']=='los-angeles']
mt_dataset = dataset[dataset['location']=='montreal']
```

```python
price_corr_sf = sf_dataset.corr()[['hourly_price_usd']].dropna()
price_corr_bt = bt_dataset.corr()[['hourly_price_usd']].dropna()
price_corr_ch = ch_dataset.corr()[['hourly_price_usd']].dropna()
price_corr_ln = ln_dataset.corr()[['hourly_price_usd']].dropna()
price_corr_la = la_dataset.corr()[['hourly_price_usd']].dropna()
price_corr_mt = mt_dataset.corr()[['hourly_price_usd']].dropna()
```

```python
#outerjoin all correlations so we can see everything
price_correlations = price_corr_ny.merge(price_corr_sf, left_index=True, right_index=True, how='outer')
price_correlations = price_correlations.merge(price_corr_bt, left_index=True, right_index=True, how='outer')
price_correlations = price_correlations.merge(price_corr_ch, left_index=True, right_index=True, how='outer')
price_correlations = price_correlations.merge(price_corr_ln, left_index=True, right_index=True, how='outer')
price_correlations = price_correlations.merge(price_corr_la, left_index=True, right_index=True, how='outer')
price_correlations = price_correlations.merge(price_corr_mt, left_index=True, right_index=True, how='outer')
price_correlations.columns=['ny_price','sf_price', 'boston_price','chicago_price','london_price', 'la_price','montreal_price']
```

```python
price_correlations['mean']=price_correlations.mean(axis='columns')
price_correlations['abs_mean']=price_correlations.mean(axis='columns').abs()
price_correlations.sort_values(by=['abs_mean'],ascending=False)
```

there is almost no correlation or rating_score with the amount of ratings. Which is weird, since reviews and rating are more likely to be done by unsatisfied customers. This will make me think that the overall client satisfaction is good. 

## Rating score correlations

```python
rat_corr_sf = sf_dataset.corr()[['rating_score']].dropna()
rat_corr_bt = bt_dataset.corr()[['rating_score']].dropna()
rat_corr_ch = ch_dataset.corr()[['rating_score']].dropna()
rat_corr_ln = ln_dataset.corr()[['rating_score']].dropna()
rat_corr_la = la_dataset.corr()[['rating_score']].dropna()
rat_corr_mt = mt_dataset.corr()[['rating_score']].dropna()
```

```python
#outerjoin all correlations so we can see everything
rat_correlations = rat_corr_ny.merge(rat_corr_sf, left_index=True, right_index=True, how='outer')
rat_correlations = rat_correlations.merge(rat_corr_bt, left_index=True, right_index=True, how='outer')
rat_correlations = rat_correlations.merge(rat_corr_ch, left_index=True, right_index=True, how='outer')
rat_correlations = rat_correlations.merge(rat_corr_ln, left_index=True, right_index=True, how='outer')
rat_correlations = rat_correlations.merge(rat_corr_la, left_index=True, right_index=True, how='outer')
rat_correlations = rat_correlations.merge(rat_corr_mt, left_index=True, right_index=True, how='outer')
rat_correlations.columns=['ny_rat','sf_rat', 'boston_rat','chicago_rat','london_rat', 'la_rat','montreal_rat']
```

```python
rat_correlations['mean']=rat_correlations.mean(axis='columns')
rat_correlations['abs_mean']=rat_correlations.mean(axis='columns').abs()
rat_correlations.sort_values(by=['abs_mean'],ascending=False)
```
